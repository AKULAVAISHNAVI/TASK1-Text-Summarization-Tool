```python
#Import Libraries
from transformers import pipeline
#use the 'summarization' pipeline, that wraps models like BART and T5 for easy use.
```

```python
#Define the Summarization Pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
#Input a Lengthy Article
long_text = """
In recent years, natural language processing (NLP) has made significant strides, thanks to the rise of deep learning and 
transformer-based models. One of the most prominent applications of NLP is automatic text summarization. The goal of text 
summarization is to condense large documents into shorter versions while preserving the core meaning. This is particularly 
useful in todays information age, where vast amounts of data are generated every second. Summarization can be broadly 
categorized into two types: extractive and abstractive. Extractive summarization involves selecting important sentences from 
the original text, while abstractive summarization generates new sentences that convey the same information. Recent models 
such as BART and T5 have shown state-of-the-art performance in abstractive summarization tasks. These models are trained on 
large corpora and are capable of generating fluent and coherent summaries. As these technologies continue to evolve, we can 
expect even more sophisticated summarization tools that better understand and interpret human language.
"""
```

```python
#Generate the Summary
summary = summarizer(long_text, max_length=100, min_length=30, do_sample=False)

# Show summary
print(" Original Article:\n", long_text)
print("\n Summary:\n", summary[0]['summary_text'])
```
